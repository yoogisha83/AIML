```python
year=range(2000,2008)
apples=[0.895,0.91,0.919,0.926,0.929,0.934,0.936,0.937] #added a value here
oranges=[0.962,0.941,0.930,0.923,0.918,0.908,0.907, 0.905] #added a value to make the lists the same length

plt.plot(year,apples)
plt.plot(year,oranges)
plt.xlabel('years')
plt.ylabel('yield')
```

```python
years = range(2000,2010)

apples=[0.895,0.91,0.919,0.926,0.929,0.934,0.936,0.937, 0.938, 0.939] # Added two values to match oranges length
oranges=[0.962,0.941,0.930,0.923,0.918,0.908,0.907,0.904,0.901, 0.9] # Added a value to match years length!

plt.plot(years,apples)
plt.plot(years,oranges)
plt.xlabel('years')
plt.ylabel('yield')
plt.title('Crop yields in Kanto')
plt.legend(['apples','oranges'])
```

```python
plt.plot(years,apples,marker='o') # Changed 'year' to 'years'
plt.plot(years,oranges,marker='x') # Changed 'year' to 'years'

plt.xlabel('years') # Added this line to label the x-axis
plt.ylabel('yield')
plt.title('Crop yields in Kanto')
plt.legend(['apples','oranges']);
```

```python
plt.plot(years,apples,marker='s',c='b',ls='-',lw=2,ms=8,mew=2,mec='navy') # Changed 'year' to 'years'
plt.plot(years,oranges,marker='o',c='r',ls='--',lw=3,ms=10,alpha=0.5) # Changed 'year' to 'years'
plt.xlabel('year')
plt.ylabel('yield')
plt.title('Crop yields in Kanto')
plt.legend(['apples','oranges'])
```

```python
plt.plot(years,apples,'s-b') # Changed 'year' to 'years'
plt.plot(years,oranges,'o--r') # Changed 'year' to 'years' and corrected typo 'orages' to 'oranges'
plt.xlabel('year')
plt.ylabel('yield')
plt.title('Crop yields in Kanto')
plt.legend(['apples','oranges'])
```

```python
import matplotlib.pyplot as plt
x=[3,1,3,12,2,4,4]
y=[3,2,1,4,5,6,7]
plt.bar(x,y)
plt.title("Bar chart")
plt.legend(["bar"])
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[3,1,3,12,2,4,4]
y=[3,2,1,4,5,6,7]
plt.scatter(x,y)
plt.legend("A")
plt.xlabel("x")
plt.ylabel("y")
plt.title("scatter chart")
plt.show()
```
